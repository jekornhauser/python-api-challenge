```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
import time

# Import API key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
humidity = []
latitude = []
longitude = []
cloudiness = []
wind_speed = []
max_temperature = []
city_name = []
city_id = []
country = []
# Set up lists for apis to pull into
#set up timer to not go over api calls per minute restriction
for i in cities:
    if cities.index(i)%50==0 and cities.index(i)!=0:
        time.sleep(90)
    # run request and if city exists in weathermap, pull the data
    url = "http://api.openweathermap.org/data/2.5/weather?q=" +i +"&units=imperial&appid=8c40b4547cf02d687f95d11b324e5504"
    print(f'Currently grabbing data about {i}.  This is number {cities.index(i)} out of {len(cities)} tested.')
    response = requests.get(url).json()
    try:
        humidity.append(response['main']['humidity'])
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        cloudiness.append(response['clouds']['all'])
        max_temperature.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
        city_name.append(response["name"])
        city_id.append(response["id"])
        country.append(response["sys"]["country"])
    except:
        print("City not found. Continuing search")
        
    
 
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#prep dictionary for dataframe conversion
dictionary = {
    "city" : city_name,
    "city_id" : city_id,
    "country" : country,
    "max_temperature" : max_temperature,
    "wind_speed" : wind_speed,
    "Cloudiness" : cloudiness,
    "Longitude" : longitude,
    "Latitude" : latitude,
    "Humidity" : humidity
}
```

```python
#convert to dataframe
weather_df= pd.DataFrame(dictionary)
weather_df.head()
```

```python
#export to .csv
weather_df.to_csv(output_data_file)
```

## Linear Regression

```python
northernhemisphere = clean_city.loc[clean_city["Latitude"]>=0]
southernhemisphere = clean_city.loc[clean_city["Latitude"]<0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.scatter(northernhemisphere["Latitude"],northernhemisphere["max_temperature"],color='blue')
(slope, intercept, rvalue, pvalue, stderr) = linregress(northernhemisphere["Latitude"], northernhemisphere["max_temperature"])
regress_values = northernhemisphere["Latitude"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(northernhemisphere["Latitude"],regress_values,"r-")
plt.annotate(line_eq,(40,62),fontsize=15,color="red")
plt.title("Northern Hemisphere City Latitude vs. Max Temperature (12/07/2020)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid()
plt.savefig("Graphs/nh_temp_and_lat.png")
```

There is a strong negative relationship between Northern Hemisphere Latitude and Max Temperature.

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.scatter(southernhemisphere["Latitude"],southernhemisphere["max_temperature"],color='blue')
(slope, intercept, rvalue, pvalue, stderr) = linregress(southernhemisphere["Latitude"], southernhemisphere["max_temperature"])
regress_values = southernhemisphere["Latitude"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(southernhemisphere["Latitude"],regress_values,"r-")
plt.annotate(line_eq,(-30,62),fontsize=15,color="red")
plt.title("Southern Hemisphere City Latitude vs. Max Temperature (12/07/2020)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid()
plt.savefig("Graphs/sh_temp_and_lat.png")
```

There is a slight positive relationship between Southern Hemisphere Latitude and Max Temperature.  This is the opposite of the Northern Relationship.  It is unexpected that it is dissimilar in correlation strength.
